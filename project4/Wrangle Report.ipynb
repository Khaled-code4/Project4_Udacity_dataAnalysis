## Introduction 

This is the fourth project of the Udacity Data Analysis Course. The project is about wrangling the data of the Twitter_archive dataset from the Twitter user @dog_rates also known as WeRateDogs.  

## Project Steps 

- step 1: Gathering data.
- step 2: Assessing data.
- step 3: Cleaning data.
- step 4: Visulization.
- step 5: Storing data.


## Gathering data 

### Twitter_archive_enhanced

There are 3 datasets we have worked on. The first one is the 'Twitter archive file' this dataset was provided by udacity and it contains all 5000+ tweets. I have downloaded it manually.

### image_archive_enhanced

The second dataset is the image prediction and it contains a prediction of the dogs. this dataset is hosted on udacity's servers and we were able to download it by using the Request library. 

### tweet_json_txt 

The last data can be gathered using Twitter API and from the tweepy library, but I couldn't so I downloaded it manually from udacity.
