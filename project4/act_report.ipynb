# WeRateDogs


WeRateDogs is a popular Twitter account known for rating and reviewing pictures of dogs submitted by users. The account rates each dog on a scale from 1 to 10 and often includes humorous commentary about the dog in the tweet. The account gained popularity for its quirky and lighthearted approach to dog ratings and has since grown into a well-known social media presence with over 9 million followers.

### The most puplouar dog stages

The four main dog stages of the WeRateDogs Twitter account are doggo, floofer, pupper, and puppo.

- Floofer: This term is used to describe a fluffy and puffy dog,
- Pupper: This term is used to describe a young or small dog
- Puppo: This term is used to describe a dog that is in between the puppy and big stage
